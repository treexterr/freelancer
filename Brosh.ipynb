```python
START_URL = 'https://brosh.kz/site_search?search_term=%09'
HEADERS = {
    'authority': 'www.olx.ua',
    'cache-control': 'max-age=0',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.106 Safari/537.36',
    'sec-fetch-dest': 'document',
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'sec-fetch-site': 'same-origin',
    'sec-fetch-mode': 'navigate',
    'sec-fetch-user': '?1',
    'accept-language': 'ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7,uk;q=0.6',
}
```

```python
def get_inside_url(search_url):
    html = requests.get(search_url, headers=HEADERS)
    soup = BeautifulSoup(html.text, 'lxml')
    if '0 наименований' in soup.text:
        print('Товар не найден')
        return False
    else:
        url = soup.find_all(class_='cs-goods-title')[0].attrs['href']
        return url
```

```python
def get_info(url):
    html = requests.get(url, headers=HEADERS)
    soup = BeautifulSoup(html.text, 'lxml')
    rrp = soup.find(class_="b-product-cost__price").text.replace('\xa0', '').replace('Тг.', '')
    img = soup.find(class_="cs-image-holder__image csjs-image").attrs['src']
    print(rrp, img)
    return rrp, img
```

```python
def make_stop(inter1, inter2):
    # Делает случайную паузу в интервале между  inter1 и inter2
    t1 = time.time()
    time.sleep(random.uniform(inter1, inter2))
    t2 = time.time()
    print(t2 - t1)
```

```python
def main():
    pricelist = {}
    with open(r'C:/Users/Эпсилон/Desktop/braslet.csv') as file:
        reader = csv.reader(file, delimiter = '^')
        for row in reader:
            pricelist[row[-1]] = row
        del pricelist['Артикул']
    for sku,value in pricelist.items():
        make_stop(0, 4)
        try:
            search_url = START_URL + sku
            print(search_url)
            url = get_inside_url(search_url)
            print(url)
            if url == False:
                rrp = 'not'
                img = 'not'
            else:
                   rrp, img = get_info(url)
            value.append(rrp)
            value.append(img)
        except:
            print('error')
            pass
    return pricelist
```
